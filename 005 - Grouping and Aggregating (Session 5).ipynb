# Grouping:

## Spliting:
    * Objects in pandas (series or Dataframe) can be split into groups by .groupby() method.

```python
# Getting data ready for split:
```

```python
pd.DatetimeIndex(s4g['Date'])
# grabs all the dates from the dataframe, s4g.
```

```python
pd.DatetimeIndex(s4g['Date']).year
# grabs all the years from the date column in the dataframe, s4g.
```

```python
s4g.insert(1, 'Year', pd.DatetimeIndex(s4g['Date']).year)
s4g
```

```python
pd.DatetimeIndex(s4g['Date']).month
```

```python
s4g.insert(2, 'Month', pd.DatetimeIndex(s4g['Date']).month)
s4g
# The date has been split into year and month so that grouping can be done on the basis of month and year for each 
# symbol, that is, stock.
```

```python
# All the four variables (Date, Year, Month, and Symbol) can be used as categorical variables. In pandas, it is 
# possible to group by any categorical variable.
```

```python
s4g
```

```python
grouped = s4g.groupby('Symbol')
grouped
# The grouping has not been performed yet and is an interim description of the grouping to be performed. This allows
# pandas to first validate that the grouping description provided to it is valid relative to the data before the 
# processing starts. Nonethless, the following operations/properties are still valid and can be performed.
```

```python
type(grouped)
```

```python
type(grouped.groups)
# the groups are of type dictionary perhaps.
```

```python
len(grouped)
# returns the number of groups which will be made once the grouping is performed.
```

```python
grouped.ngroups
# I think it returns the same output as the len() method does.
```

```python
grouped.groups
# returs a dictionary where keys are names of the groups, and the values are the indexes contained in each group.
# In this case, the names of the groups are AAPL and MSFT and the values are the indices in each group.
```

```python
grouped
```

```python
def print_groups(groupobject):
    for name, group in groupobject:
        print (name)
        print (group)
# to retreive group names from by-group objects.
# it also shows that how the data will be grouped.
```

```python

print_groups(grouped)
# Pandas takes the distinct values from the groupby column and creates a seperate dataframe for each distinct value. It then copies
# the non-grouped columns and data into each dataframe and call the groups by the distinct values.
```

```python
grouped.size()
# outputs the size of each group (number of rows in each group)
```

```python
grouped.get_group('MSFT')
# outputs the required group.
```

```python
mcg = s4g.groupby(['Symbol', 'Year', 'Month'])
mcg
# First the entire dataframe (s4g) will be grouped into two dataframes based on Symbol. Then each dataframe is further grouped
# by year, but since there is only one year, 2012, therefore, it has no effect. That is, the total number of grouped dataframes
# are still 2. Finally, each dataframe is further grouped by months, and there are 12 months, so each dataframe is further
# grouped into 12 dataframes (groups). Thus, the total number of groups made are 12*2=24.
```

```python
type(mcg)
# mcg is a dataframe.
```

```python
type(mcg.groups)
# the groups in mcg are dictionaries.
```

```python
mcg.groups
```

```python
print_groups(mcg)
# 12 groups formed for AAPL and 12 groups formed for MSFT. Total 24 groups formed since the year is only 2012 and 12 months are in
# the year.
```

## Aggregating:

```python
mig_112.agg(np.mean)
# mig_112 has 24 groups as seen above. This command will create mean for each group so in total we should have 24 means.
# The output has the identically structured indices as the original data, mi. As the original data, mi, the following
# output of aggregation has three indices (Symbol, Year, and Month).
```

```python
type(mig_112.agg(np.mean))
```

```python
mig_112.agg(np.mean).loc['AAPL'][3:7]
```

```python
s4g
# the difference between s4g and mi is that s4g has zero-based integer index, whereas mi has three indices (Symbol, 
# Year, and Month).
```

```python
test = s4g.groupby(['Symbol', 'Year', 'Month'], as_index=False)
test
# This output of aggregation is based on zero-based integer, rather than 3 indices ('Symbol', 'Year', and 'Month') 
# unlike the aggregation output generated above. The main difference is this aggregation is based on s4g, which has
# zero-based integer index, whereas the aggregation output generated above is based on mi, which has 3 indices ('Symbol,
# Year, and Month'). Apart from this index differences, the output of both aggregations is the same, that is, the same
# group means have been obtained in both cases (when the aggregation is based on s4g and on mi).
```
